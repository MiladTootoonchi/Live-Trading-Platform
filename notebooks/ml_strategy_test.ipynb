# Testing ML_strategy function
This notebook is only going to test if the ml_strategy function returns expected values.

### Importing

```python
import sys
from pathlib import Path
import requests
import asyncio

sys.path.append(str(Path("..").resolve()))

from src.ml_model.ml_strategy import AI_strategy
from config import load_api_keys
```

### Getting positions

```python
# Loading keys
key, secret_key = load_api_keys() 

HEADERS = {
        "APCA-API-KEY-ID" : key,
        "APCA-API-SECRET-KEY": secret_key,
}

APCA_API_BASE_URL =  "https://paper-api.alpaca.markets"

url = f"{APCA_API_BASE_URL}/v2/positions"
response = await asyncio.to_thread(requests.get, url, headers = HEADERS)
positions = response.json()
```

### Checking the signals and qty for each position

```python
for position in positions:
    symbol_i = position.get("symbol")
    signal, qty = await AI_strategy(position)
    
    print(f"{symbol_i}: the AI says you should {signal} with qty = {qty}")
```
